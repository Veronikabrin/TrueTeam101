```python
def Vectorizer_model():
    import pandas as pd 
    import re
    data = pd.read_csv('1.4.csv', error_bad_lines = False, encoding = "utf-8")
    a = list(data['tweets'])
    from sklearn.feature_extraction.text import TfidfVectorizer
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform(a[:500])
    feature_names = vectorizer.get_feature_names()
    dense = vectors.todense()
    denselist = dense.tolist()
    df = pd.DataFrame(denselist, columns=feature_names)
    return [df, feature_names]
def clusters(df):
    import hdbscan
    from sklearn.manifold import TSNE
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    projection = TSNE().fit_transform(df)
    clusterer = hdbscan.HDBSCAN(min_cluster_size=4, prediction_data=True).fit(df)
    return clusterer
def save_data():
    import pickle
    l = Vectorizer_model()
    df = l[0]
    columns = l[1]
    f = open( 'columns.txt', 'w')
    f.write(' '.join(columns))
    f.close()
    c = clusters(df)
    pickle.dump(c, open('finalized_model.sav', 'wb'))
    import csv
    df.to_csv('df.csv')
def cleane_msg(msg):
    import re
    reg = re.compile('[^а-яА-Я ]')
    for i in range(len(msg)):
        msg[i] = re.sub(r'[,.;?!)(]', ' ', msg[i])
        msg[i] = reg.sub('', msg[i])
        msg[i] = msg[i].lower()
        a = msg[i].split()
        newlist =[]
        for j in a: 
            if j not in newlist: newlist.append(j) 
        if len(newlist) > 1 : msg[i] = ' '.join(newlist)
    msg = list(filter(lambda x: len(x) > 3, msg))  
    return msg
def Vectorizer(text):
    import pandas as pd 
    import re
    from sklearn.feature_extraction.text import TfidfVectorizer
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform(list(text))
    feature_names = vectorizer.get_feature_names()
    dense = vectors.todense()
    denselist = dense.tolist()
    df = pd.DataFrame(denselist, columns=feature_names)
    return df
def prediction(text):
    import hdbscan
    import pickle
    import pandas as pd
    from pymystem3 import Mystem
    m = Mystem()
    m.start()
    lemmas = m.lemmatize(text)
    lemmas = cleane_msg(lemmas)
    clusterer = pickle.load(open('finalized_model.sav', 'rb'))
    f = open( 'columns.txt', 'r')
    columns = f.read()
    columns = columns.split()
    test = Vectorizer(lemmas)
    test_f = pd.DataFrame(columns = columns)
    for i in range(len(lemmas)):
        test_f.loc[i] = 0
        for j in test.columns:
            if j in test_f.columns:
                test_f[j][i] = test[j][i]
                
    predict_class, strengths = hdbscan.approximate_predict(clusterer, test_f)
    
    data = pd.read_csv('1.4.csv', error_bad_lines = False, encoding = "utf-8")
    data=data[:500]
    data['probabilities'] = clusterer.probabilities_
    data['labels'] = clusterer.labels_
    
    
    #from gensim.summarization.summarizer import summarize
    #text_of_class = summarize((data.loc[data['labels'] ==predict_class]), ratio=0.01)
    return predict_class


#save_data() #юзать надо только один раз чтоб была модель
test = 'мочwerь'
#!pip install gensim
print(prediction(test))
```
